#### Q1. Read a file constitution.txt. Create a dictionary with an entry for each letter in the alphabet. The keys are letters and the values will be the counts of the number of times a letter has viewed. Go through each character in data, skipping characters that aren't letters. For every letter, increment the count stored in letter_counts for that letter. Create a bar chart for the letter frequencies

```python
import pandas as pd

def letterCounter (demo_str):
    
    alpha_dict = {}
    for letter in demo_str:
        if letter.isalpha():
            if letter in alpha_dict:
                alpha_dict[letter] += 1
            else:
                alpha_dict[letter] = 1

    return alpha_dict

file = open('constitution.txt', 'r')
text = file.read().lower()

disp_graph= pd.Series(letterCounter(text))
disp_graph.plot.bar(figsize=(20,5))
```
